```python
import string

# 建立字母數字映射
alphabet = string.ascii_lowercase
letter_to_number = {letter: number for letter, number in zip(alphabet, range(1, 27))}

# 測試轉換
input_letter = 'a'
if input_letter in letter_to_number:
    corresponding_number = letter_to_number[input_letter]
    print(f"The letter '{input_letter}' corresponds to the number {corresponding_number}.")
else:
    print(f"'{input_letter}' is not a valid lowercase letter.")

# 轉換整個單詞
word = 'hello'
number_list = [str(letter_to_number[letter]) for letter in word]
result = ', '.join(number_list)
print(f"The word '{word}' corresponds to the numbers: {result}")
```
